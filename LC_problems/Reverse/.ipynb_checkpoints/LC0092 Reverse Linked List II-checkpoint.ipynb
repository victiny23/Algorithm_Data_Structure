Given the head of a singly linked list and two integers left and right where left <= right, reverse the nodes of the list from position left to position right, and return the reversed list.

 

Example 1:


Input: head = [1,2,3,4,5], left = 2, right = 4
Output: [1,4,3,2,5]
Example 2:

Input: head = [5], left = 1, right = 1
Output: [5]
 

Constraints:

The number of nodes in the list is n.
1 <= n <= 500
-500 <= Node.val <= 500
1 <= left <= right <= n

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:        
    def reverseBetween(self, head: ListNode, left: int, right: int) -> ListNode:
        dummy = ListNode(0)
        dummy.next = head
        p = dummy
        for _ in range(left - 1):
            p = p.next
        
        prev = None
        curr = p.next
        for _ in range(right - left + 1):
            foll = curr.next
            curr.next = prev
            prev = curr
            curr = foll
        p.next.next = curr
        p.next = prev
        return dummy.next
```

```python
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prev, curr = None, head
        # the linked-list is broke into 2 separate linked list
        # one led by prev, the other lead by curr 
        while curr:
            foll = curr.next
            curr.next = prev
            prev = curr
            curr = foll
        return prev
        
    def reverseBetween(self, head: ListNode, left: int, right: int) -> ListNode:
        if not head.next or left == right:
            return head
        dummy = ListNode(0)
        dummy.next = head
        lpre = dummy
        for _ in range(left - 1):
            lpre = lpre.next
        lcur = lpre.next
        rpre = lcur
        for _ in range(right - left):
            rpre = rpre.next
        rcur = rpre.next
        rpre.next = None
        lpre.next = self.reverseList(lcur)
        lcur.next = rcur
        return dummy.next
```
