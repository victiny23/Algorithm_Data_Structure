Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the frequency of at least one of the chosen numbers is different.

It is guaranteed that the number of unique combinations that sum up to target is less than 150 combinations for the given input.

 

Example 1:

Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.
Example 2:

Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]
Example 3:

Input: candidates = [2], target = 1
Output: []
Example 4:

Input: candidates = [1], target = 1
Output: [[1]]
Example 5:

Input: candidates = [1], target = 2
Output: [[1,1]]
 

Constraints:

1 <= candidates.length <= 30
1 <= candidates[i] <= 200
All elements of candidates are distinct.
1 <= target <= 500


```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        """Backtracking: update currSum - 113 ms"""
        def dfs(candidates, currSum, combo, res, target):
            if currSum > target:
                return # backtracking
            if currSum == target:
                res.append(combo)
                return
            for i in range(len(candidates)):
                dfs(candidates[i:], currSum + candidates[i], combo + [candidates[i]], res, target)
        res = []
        dfs(candidates, 0, [], res, target)
        return res
```

```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        """Sort + Dynamic Programming - 101 ms"""
        candidates.sort()
        dp = [[] for _ in range(target + 1)]
        for t in range(1, target + 1): # each possible value <= target
            for candi in candidates: # each candidate in candidates
                if candi > t:
                    break
                if candi == t:
                    dp[t].append([candi])
                # dp[t - candi] are the combinations with sum equal to t - candi
                # check candi >= combo[-1] to avoid duplicated answers like [3,2,2], [2,3,2], [2,2,3]
                dp[t] += [combo + [candi] for combo in dp[t - candi] if candi >= combo[-1]]
        return dp[target]
```

```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]: 
        """Dynamic Programming, No Sorting - 91 ms"""
        # use 3d matrix to store the combinations
        # dp[t][i]: the ith combination for sum equals to t
        dp = [[[]]] + [[] for _ in range(target)]
        # initial dp: [[[]], [], [], [], [], [], [], []]
        # [[]] is the combination for target == 0, + empty space for combinations ranging from 1 to target
        for candi in candidates:  # for each candidate
            for t in range(candi, target + 1):  # candi <= each possible value <= target
                # dp[t - candi] are the combinations with sum equal to t - candi
                dp[t] += [combo + [candi] for combo in dp[t - candi]]          
        return dp[target]
```
