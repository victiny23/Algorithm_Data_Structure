Given a string s and a string array dictionary, return the longest string in the dictionary that can be formed by deleting some of the given string characters. If there is more than one possible result, return the longest word with the smallest lexicographical order. If there is no possible result, return the empty string.

 

Example 1:

Input: s = "abpcplea", dictionary = ["ale","apple","monkey","plea"]
Output: "apple"
Example 2:

Input: s = "abpcplea", dictionary = ["a","b","c"]
Output: "a"
 

Constraints:

1 <= s.length <= 1000
1 <= dictionary.length <= 1000
1 <= dictionary[i].length <= 1000
s and dictionary[i] consist of lowercase English letters.

```python
class Solution:
    def findLongestWord(self, s: str, dictionary: List[str]) -> str:
        """sort the word in the dictionary and then check if word is in s - for early exit - 208 ms"""
        sorted_dic = sorted(dictionary, key = lambda x: (-len(x), x))
        i = 0
        for word in sorted_dic:
            # check if the word is in sorted dictionary
            i = 0
            for c in s:
                if c == word[i]:
                    i += 1
                    # if such a word is found
                    if i == len(word):
                        return word
        return ""
```

```python
class Solution:
    def findLongestWord(self, s: str, dictionary: List[str]) -> str:
        """check if word is in s + compare length - 376 ms"""
        res = ""
        
        def check(s, w):
            i = 0
            for c in s:
                if c == w[i]:
                    i += 1
                # reach the end of word: current word found in s
                if i == len(w):
                    return True
            return False
        
        for word in dictionary:
            # check if word is in s
            if check(s, word):
                # compare the length
                if len(word) > len(res) or (len(word) == len(res) and word < res):
                    res = word
                # # another way for comparison
                # res = min(res, word, key = lambda x: (-len(x), x))
        return res
```
