The Fibonacci numbers, commonly denoted F(n) form a sequence, called the Fibonacci sequence, such that each number is the sum of the two preceding ones, starting from 0 and 1. That is,

F(0) = 0, F(1) = 1
F(n) = F(n - 1) + F(n - 2), for n > 1.
Given n, calculate F(n).

 

Example 1:

Input: n = 2
Output: 1
Explanation: F(2) = F(1) + F(0) = 1 + 0 = 1.
Example 2:

Input: n = 3
Output: 2
Explanation: F(3) = F(2) + F(1) = 1 + 1 = 2.
Example 3:

Input: n = 4
Output: 3
Explanation: F(4) = F(3) + F(2) = 2 + 1 = 3.
 

Constraints:

0 <= n <= 30

```python
class Solution:
    def fib(self, n: int) -> int:
        """bottom-up: O(n) space"""
        if n == 0:
            return 0
        if n == 1:
            return 1
        f = [0, 1]
        for i in range(n-1):
            f.append(f[-1] + f[-2])
        return f[-1]
        
        
        """bottom-up: O(1) sapce"""
        if n == 0:
            return 0
        if n == 1:
            return 1
        a, b = 0, 1
        for i in range(n):
            a, b = b, a + b
        return a
        
        
        """top-down"""
        if n == 0:
            return 0
        if n == 1:
            return 1
        return self.fib(n-1) + self.fib(n-2)
    
    
        """top-down + memorization (list)"""
        if n == 0:
            return 0
        if n == 1:
            return 1
        li = [-1 for _ in range(n+1)]
        li[0], li[1] = 0, 1
        def helper(n, li):
            if li[n] < 0:
                li[n] = helper(n-1, li) + helper(n-2, li)
            return li[n]
        return helper(n, li)
        
        
        """top-down + memorization (dict)"""
        if n == 0:
            return 0
        if n == 1:
            return 1
        dic = {0: 0, 1: 1}
        def helper(n, dic):
            if n not in dic:
                dic[n] = helper(n-1, dic) + helper(n-2, dic)
            return dic[n]
        return helper(n, dic)
        
        
        """math: using the Fibonacci formula 1"""
        root5 = pow(5, 0.5)
        return round((pow((1+root5)/2, n) - pow((1-root5)/2, n))/root5)
    
    
        """math: using the Fibonacci formula 2"""
        phi = (1 + pow(5, 0.5))/2
        return round(pow(phi, n)/pow(5, 0.5))
```
