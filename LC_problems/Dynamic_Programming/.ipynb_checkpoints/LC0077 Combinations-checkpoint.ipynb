Given two integers n and k, return all possible combinations of k numbers out of the range [1, n].

You may return the answer in any order.

 

Example 1:

Input: n = 4, k = 2
Output:
[
  [2,4],
  [3,4],
  [2,3],
  [1,2],
  [1,3],
  [1,4],
]
Example 2:

Input: n = 1, k = 1
Output: [[1]]
 

Constraints:

1 <= n <= 20
1 <= k <= n

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        """itertools.combinations - 78 ms"""
        from itertools import combinations
        return combinations(range(1, n+1), k)

        
       
        
        
        
       
        
        
```

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        """Recursive - 165 ms"""
        if k == 0:
            return [[]]
        # because there are k - 1 numbers in pre
        # num must starts from k
        return [pre + [num] for num in range(k, n+1) for pre in self.combine(num-1, k-1)]
```

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        """Iterative - 801 ms"""
        res = [[]]
        # iterate k times to update res
        # smaller numbers are added to the left of the path from a previous round
        for _ in range(k):
            res = [[i] + path for path in res for i in range(1, path[0] if path else n+1)]
        return res
```

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
         """Backtracking - 451 ms"""
        def dfs(remain, path):
            # if remain < 0:
            #     return # backtracking
            if remain == 0:
                res.append(path)
                return
            if not path:
                for i in range(1, n + 1):
                    dfs(remain - 1, [i])
            else:
                for i in range(path[-1] + 1, n + 1):
                    dfs(remain - 1, path + [i])
        res = []
        dfs(k, [])
        return res
```

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        """Backtracking - updating the list of available numbers - 663 ms"""
        def dfs(li, remain, path):
            if remain == 0:
                res.append(path)
                return
            for i in range(len(li)):
                dfs(li[i+1:], remain - 1, path + [li[i]])
        res = []
        dfs(range(1, n + 1), k, [])
        return res
        
        """Backtracking - updating the list of available numbers - 740 ms"""
        def dfs(li, path):
            if len(path) == k:
                res.append(path)
                return
            for i in range(len(li)):
                dfs(li[i+1:], path + [li[i]])
        res = []
        dfs(range(1, n + 1), [])
        return res
```

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
         """DP - 1105 ms"""
        dp = [[] for _ in range(k)]
        dp[0] = [[i] for i in range(1, n+1)]
        for i in range(1, k):
            for path in dp[i-1]:
                for num in range(path[-1]+1, n+1):
                    dp[i].append(path + [num])
        return dp[-1]
```

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        """Stack:  - 114 ms
        keep pushing nembers into the stack until length is met, 
        then vary the top of the stack to find different combinations,
        then extend the varying part to the left until the bottom of the stack"""
        res = []
        stack = []
        x = 1
        while True:
            l = len(stack)
            # a combination is found
            if l == k:
                res.append(stack[:])
            # l is the length of the current path, we need k - l more numbers
            # there are n - (x-1) numbers from x to n
            # if k - l > n - (x-1), there are not enough number to make the combination
            # when a combination is formed, or no combination can be found, check for return
            if l == k or k - l > n - (x - 1):
                # empty stack, no more available number for a potential combination
                if not stack:
                    return res
                # case 1: l == k, a combination is found, pop x and try x + 1
                # case 2: k - l > n - (x - 1): x or greater wont' work, x is not appended
                #         stack top (y < x) is popped, we try y + 1 in the next round
                x = stack.pop() + 1
            else:
                # l < k and there are enough numbers available in the rest of the numbers
                # we can push x into the stack for a potwntial combination 
                stack.append(x)
                x += 1
```
