You are climbing a staircase. It takes n steps to reach the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

 

Example 1:

Input: n = 2
Output: 2
Explanation: There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps
Example 2:

Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
 

Constraints:

1 <= n <= 45

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        """the last step comes from (n-1)th step or (n-2)th step"""
        """dynamic programming"""
        
        """bottom-up, O(n) space"""
        if n == 1:
            return 1
        if n == 2:
            return 2
        steps = [1, 2]
        for i in range(n-2):
            steps.append(steps[-1] + steps[-2])
        return steps[-1]
    
        
        """bottom-up, O(1) space"""
        a, b = 1, 1
        for i in range(n-1):
            a, b = b, a + b
        return b
    
    
        """top-down - TLE"""
        if n == 1:
            return 1
        if n == 2:
            return 2
        return self.climbStairs(n-1)+self.climbStairs(n-2)
    
        
        """top-down + memorization (list)"""
        if n == 1:
            return 1
        if n == 2:
            return 2
        li = [0 for i in range(n)]
        li[0], li[1] = 1, 2
        def helper(n, li):
            # check if the result is already in the list
            if not li[n]:
                li[n] = helper(n-1, li) + helper(n-2, li)
            return li[n]
        return helper(n-1, li)
    
        
        """top-down + memorization (dictionary)"""
        if n == 1:
            return 1
        if n == 2:
            return 2
        dic = {1: 1, 2: 2}
        def helper(n, dic):
            # check if the result is already in the dict
            if n not in dic:
                dic[n] = helper(n-1, dic) + helper(n-2, dic)
            return dic[n]
        return helper(n, dic)
    
        
        """math: using the Fibonacci formula 1"""
        n += 1
        root5 = pow(5, 0.5)
        return round((pow((1+root5)/2, n) - pow((1-root5)/2, n))/root5)
        
    
        """math: using the Fibonacci formula 2"""
        phi = (1 + pow(5, 0.5))/2
        return round(pow(phi, n+1)/pow(5, 0.5))
```
