Given an array of integers nums, half of the integers in nums are odd, and the other half are even.

Sort the array so that whenever nums[i] is odd, i is odd, and whenever nums[i] is even, i is even.

Return any answer array that satisfies this condition.

 

Example 1:

Input: nums = [4,2,5,7]
Output: [4,5,2,7]
Explanation: [4,7,2,5], [2,5,4,7], [2,7,4,5] would also have been accepted.
Example 2:

Input: nums = [2,3]
Output: [2,3]
 

Constraints:

2 <= nums.length <= 2 * 104
nums.length is even.
Half of the integers in nums are even.
0 <= nums[i] <= 1000
 

Follow Up: Could you solve it in-place?

```python
class Solution:
    def sortArrayByParityII(self, nums: List[int]) -> List[int]:
        """divide nums into 2 lists odd and even, then combine, O(n) space"""
        even, odd = [], []
        while nums:
            num = nums.pop()
            if num % 2 == 0:
                even.append(num)
            else:
                odd.append(num)
        li = [0] * 2 * len(even)
        li[::2] = even
        li[1::2] = odd
        return li
        
        """divide nums into 2 lists odd and even, then combine, O(n) space"""
        even, odd = [], []
        while nums:
            num = nums.pop()
            if num % 2 == 0:
                even.append(num)
            else:
                odd.append(num)
        li = []
        while even:
            li += [even.pop(), odd.pop()]
        return li  
```

```python
class Solution:
    def sortArrayByParityII(self, nums: List[int]) -> List[int]:
        """two pointer with O(n) space"""
        li = [0] * len(nums)
        i, j = 0, 1
        for num in nums:
            if num % 2 == 0:
                li[i] = num
                i += 2
            else:
                li[j] = num
                j += 2
        return li
```

```python
class Solution:
    def sortArrayByParityII(self, nums: List[int]) -> List[int]:
        """
        two pointer from head, one for even index and one for odd index, both with step 2, O(1) space
        checking num at both pointer
        """
        i, j = 0, 1
        n = len(nums)
        while i < n and j < n:
            # even index being even checked
            if nums[i] % 2 == 0:
                i += 2
            # odd index being odd checked
            elif nums[j] % 2 == 1:
                j += 2
            # even index being odd and odd index being even - swap to meet the requirement
            else:
                nums[i], nums[j] = nums[j], nums[i]
                i += 2
                j += 2
        return nums
        
        """
        two pointer from head, one for even index and one for odd index, both with step 2, , O(1) space
        checking num at only one pointer - swap would meet the requirement for the other pointer
        """
        i, j = 0, 1
        n = len(nums)
        while j < n:
            # even number found at odd index, move it to an even index
            # update the pointer for even index and check the currrent location again
            if nums[j] % 2 == 0:
                nums[i], nums[j] = nums[j], nums[i]
                i += 2
            else:
                j += 2
        return nums
```

```python
class Solution:
    def sortArrayByParityII(self, nums: List[int]) -> List[int]:
        """
        3 pointer, 
        one head (with step 1) for the num to check, 
        two tail (with step 2) for odd and even index,
        O(1) space
        """
        i, j, k = 0, len(nums)-2, len(nums)-1
        while i <= j or i <= k:
            if i % 2 == 0:
                if nums[i] % 2 == 1:
                    nums[i], nums[k] = nums[k], nums[i]
                    k -= 2
                else:
                    i += 1
            else:
                if nums[i] % 2 == 0:
                    nums[i], nums[j] = nums[j], nums[i]
                    j -= 2
                else:
                    i += 1
        return nums
```
