Given two non-negative integers num1 and num2 represented as strings, return the product of num1 and num2, also represented as a string.

Note: You must not use any built-in BigInteger library or convert the inputs to integer directly.

 

Example 1:

Input: num1 = "2", num2 = "3"
Output: "6"
Example 2:

Input: num1 = "123", num2 = "456"
Output: "56088"
 

Constraints:

1 <= num1.length, num2.length <= 200
num1 and num2 consist of digits only.
Both num1 and num2 do not contain any leading zero, except the number 0 itself.

```python
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        """assign a list of length n1 * n2"""
        res = [0] * (len(num1) + len(num2))
        for i in range(len(num1)-1, -1, -1):
            n1 = ord(num1[i]) - ord('0')
            for j in range(len(num2)-1, -1, -1):
                n2 = ord(num2[j]) - ord('0')
                tmp = n1 * n2 + res[i+j+1] # i <= len1 -1, j <= len2 -1, the last digit is on len1 + len2 - 1
                res[i+j+1] = tmp % 10
                res[i+j] += tmp // 10 # push to the digit on the left
        # skip the leading 0's
        for i, d in enumerate(res):
            # find the 1st non-0 digit
            if d != 0:
                return ''.join(map(str, res[i:]))
        # string of all 0's
        return '0'
```

```python
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        if num1 == '0' or num2 == '0':
            return '0'
        def multiply_digit(num, digit):
            res = []
            carry = 0
            l = len(num)
            while l or carry:
                if l:
                    carry += (ord(num[l-1]) - ord('0')) * digit
                    l -= 1
                res.append(carry % 10)
                carry //= 10
            return ''.join(map(str, res))[::-1]
        
        def add_nums(num1, num2):
            l1, l2 = len(num1), len(num2)
            res = []
            carry = 0
            while l1 or l2 or carry:
                if l1:
                    carry += ord(num1[l1-1]) - ord('0')
                    l1 -= 1
                if l2:
                    carry += ord(num2[l2-1]) - ord('0')
                    l2 -= 1
                res.append(carry % 10)
                carry //= 10
            return ''.join(map(str, res))[::-1]
        
        n1, n2 = len(num1), len(num2)
        if n1 < n2:
            self.multiply(num2, num1)
        res ='0'
        for i in range(n2):
            prod = multiply_digit(num1, ord(num2[i]) - ord('0'))
            prod += '0' * (n2 - i - 1) # if prod == 0, this could give leading 0 like '0000'
            res = add_nums(res, prod)
        return res
```
