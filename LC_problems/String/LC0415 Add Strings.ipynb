Given two non-negative integers, num1 and num2 represented as string, return the sum of num1 and num2 as a string.

You must solve the problem without using any built-in library for handling large integers (such as BigInteger). You must also not convert the inputs to integers directly.

 

Example 1:

Input: num1 = "11", num2 = "123"
Output: "134"
Example 2:

Input: num1 = "456", num2 = "77"
Output: "533"
Example 3:

Input: num1 = "0", num2 = "0"
Output: "0"
 

Constraints:

1 <= num1.length, num2.length <= 104
num1 and num2 consist of only digits.
num1 and num2 don't have any leading zeros except for the zero itself.

```python
class Solution:
    def addStrings(self, num1: str, num2: str) -> str:
        """ord(x) returns the unicode code point of character x. 
        By doing, ord(x) - ord('0'), we will get the numerical value of x 
        since the unicode code points for 0, 1, ..., 9 are also in ascending order. """
        num1 = list(num1)
        num2 = list(num2)
        num1 = [int(i) for i in num1]
        num2 = [int(i) for i in num2]
        carry = 0
        res = []
        while num1 and num2:
            digit = ord(num1.pop()) - ord('0') + ord(num2.pop()) - ord('0') + carry
            carry = digit // 10
            res.append(digit%10)
        while num1:
            digit = ord(num1.pop()) - ord('0') + carry
            carry = digit // 10
            res.append(digit%10)
        while num2:
            digit = ord(num2.pop()) - ord('0') + carry
            carry = digit // 10
            res.append(digit%10)
        if carry:
            res.append(carry)
        return ''.join(map(str, res[::-1]))
        
        """ord('0')"""
        num1, num2 = list(num1), list(num2)
        carry = 0
        res = []
        while num1 or num2:
            n1, n2 = 0, 0
            if num1: 
                n1 = ord(num1.pop()) - ord('0')
            if num2: 
                n2 = ord(num2.pop()) - ord('0')
            carry, remain = divmod(n1 + n2 + carry, 10)
            res.append(remain)
        if carry:
            res.append(carry)
        return ''.join(map(str, res))[::-1]
        
        """not converting to lists"""
        l1, l2 = len(num1), len(num2)
        carry = 0
        res = []
        while l1  or l2  or carry:
            if l1:
                carry += ord(num1[l1-1]) - ord('0')
                l1 -= 1
            if l2:
                carry += ord(num2[l2-1]) - ord('0')
                l2 -= 1
            res.append(carry%10)
            carry //= 10
        return ''.join(map(str, res))[::-1]
            
```
