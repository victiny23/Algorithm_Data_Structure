Given the API rand7() that generates a uniform random integer in the range [1, 7], write a function rand10() that generates a uniform random integer in the range [1, 10]. You can only call the API rand7(), and you shouldn't call any other API. Please do not use a language's built-in random API.

Each test case will have one internal argument n, the number of times that your implemented function rand10() will be called while testing. Note that this is not an argument passed to rand10().

Follow up:

What is the expected value for the number of calls to rand7() function?
Could you minimize the number of calls to rand7()?
 

Example 1:

Input: n = 1
Output: [2]
Example 2:

Input: n = 2
Output: [2,8]
Example 3:

Input: n = 3
Output: [3,8,10]
 

Constraints:

1 <= n <= 105

```python
# The rand7() API is already defined for you.
# def rand7():
# @return a random integer in the range 1 to 7

class Solution:
    def rand10(self):
                """
        :rtype: int
        """
        """Rejection Sampling
        Increase the range of the unifrom distribution, then sample from the extended distribution"""    
        """1"""
        def rand49():
            # divide a multiple of 7 into several sub-domains
            # 1st half determines the sub-domain uniformly
            # 2nd hald determines the location in the sub-domain
            return (rand7() - 1)* 7 + rand7()
        
        while True:
            # Increase the range of the unifrom distribution
            i = rand49()
            # values from 41 - 49 do not make a unifrom distribution because the lack of 50
            # sample from the extended distribution
            if i <= 40:
                return 1 + i % 10
        
        """2"""
        while True:
            i = (rand7() - 1)* 7 + rand7()
            if i <= 40:
                return 1 + i % 10
        
        """3"""
        randx = 41
        while randx > 40:
            randx = (rand7() - 1)* 7 + rand7()
        return 1 + randx % 10
        
        """4"""
        i = (rand7() - 1)* 7 + rand7()
        return 1 + i % 10 if i <= 40 else self.rand10() 
```

```python
class Solution:
    def rand10(self):
        """Split the possible value into 2 groups and select by odd/even"""
        group = 7
        while group > 6:
            group = rand7() # group can be 1, 2, 3, 4, 5, 6 with equal probability
        value = 6
        while value > 5:
            value = rand7() # value can be 1, 2, 3, 4, 5 with equal probability
        # randomize the group
        if group % 2 == 0:
            return value
        else:
            return value + 5
```
