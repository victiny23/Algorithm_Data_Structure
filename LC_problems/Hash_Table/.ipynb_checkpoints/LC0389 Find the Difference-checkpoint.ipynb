You are given two strings s and t.

String t is generated by random shuffling string s and then add one more letter at a random position.

Return the letter that was added to t.

 

Example 1:

Input: s = "abcd", t = "abcde"
Output: "e"
Explanation: 'e' is the letter that was added.
Example 2:

Input: s = "", t = "y"
Output: "y"
Example 3:

Input: s = "a", t = "aa"
Output: "a"
Example 4:

Input: s = "ae", t = "aea"
Output: "a"
 

Constraints:

0 <= s.length <= 1000
t.length == s.length + 1
s and t consist of lower-case English letters.

```python
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        """Bit Operation XOR: exclusive or"""
        x = 0
        for ss in s:
            x ^= ord(ss) # character to int
        for tt in t:
            x ^= ord(tt) # int to character
        return chr(x)
```

```python
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        """Sorting"""
        ls = [_ for _ in s]
        lt = [_ for _ in t]
        ls.sort()
        lt.sort()
        for i in range(len(ls)):
            if ls[i] != lt[i]:
                return lt[i]
        return lt[-1]
```

```python
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        """Hash Table"""
        # 1 add all from t then remove all from s, what's left is the extra
        dic = {}
        for tt in t:
            dic[tt] = dic.get(tt, 0) + 1
        for ss in s:
            dic[ss] = dic.get(ss, 0) - 1
            if dic[ss] == 0:
                del dic[ss]
        return list(dic.keys())[0]
        
        # 2 add all from s then remove elements from t, what's not found is the extra
        dic = {}
        for ss in s:
            dic[ss] = dic.get(ss, 0) + 1
        for tt in t:
            if tt not in dic.keys():
                return tt
            dic[tt] -= 1
            if dic[tt] == 0:
                del dic[tt]    
```
