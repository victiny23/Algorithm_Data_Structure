Given an array of strings words representing an English Dictionary, return the longest word in words that can be built one character at a time by other words in words.

If there is more than one possible answer, return the longest word with the smallest lexicographical order. If there is no answer, return the empty string.

 

Example 1:

Input: words = ["w","wo","wor","worl","world"]
Output: "world"
Explanation: The word "world" can be built one character at a time by "w", "wo", "wor", and "worl".
Example 2:

Input: words = ["a","banana","app","appl","ap","apply","apple"]
Output: "apple"
Explanation: Both "apply" and "apple" can be built from other words in the dictionary. However, "apple" is lexicographically smaller than "apply".
 

Constraints:

1 <= words.length <= 1000
1 <= words[i].length <= 30
words[i] consists of lowercase English letters.

```python
class Solution:
    def longestWord(self, words: List[str]) -> str:
        """ 68 ms
        Set + sort the word from short to long
        build the word from an emoty string:
        if word[:-1] in the set, add word to the set - update longest_word if word added is longer
        the set has all valid words
        """
        words.sort()
        words_set, res = set([""]), ""
        for word in words:
            if word[:-1] in words_set:
                # another valid word  is added to the set
                words_set.add(word)
                # word of the same length with larger lexicographical order come up later in sorted words
                # they won't be used to update res as they are not longer than the smaller word
                if len(word) > len(res):
                    res = word   
        return res   
```

```python
class Solution:
    def longestWord(self, words: List[str]) -> str:
        """ 108 ms
        Sort word by length from long to short, 
        slice the word to check if the remainder in words: 
        if not found - next word
        if all slices found - return word"""
        sorted_w = sorted(words, key=lambda x: (-len(x), x))
        
        res = ""
        for i, word in enumerate(sorted_w):
            dic = [word[:i] for i in range(1, len(word))]
            for j in range(len(dic)):
                if dic[j] in sorted_w[i+1:]:
                    j += 1
                else:
                    break
            if j == len(dic):
                return word
        return res
```

```python
class Solution:
    def longestWord(self, words: List[str]) -> str:
        """ 124 ms
        BFS: start from a queue of an empty string
        + a Set of all words
        pop a prev word from the queue
        search through ascii_lowercase(i.e. a-z) to see if prev + any letter makes a new word in the Set
        if so - add the new word to the queue, remove it from the Set
        if not - word is the longest with the current beginings
        """
        words_set = set(words)
        queue = deque([""])
        res = ""
        while queue:
            # previous word found in set
            prev = queue.popleft()
            # update the result if necessary
            if len(prev) > len(res):
                res = prev
            # extend the prev to see if an extended word can be found in the set
            for c in ascii_lowercase:
                extend = prev + c
                if extend in words_set:
                    queue.append(extend)
                    words_set.remove(extend)
        return res
```
