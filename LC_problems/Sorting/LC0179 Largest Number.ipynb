Given a list of non-negative integers nums, arrange them such that they form the largest number.

Note: The result may be very large, so you need to return a string instead of an integer.

 

Example 1:

Input: nums = [10,2]
Output: "210"
Example 2:

Input: nums = [3,30,34,5,9]
Output: "9534330"
Example 3:

Input: nums = [1]
Output: "1"
Example 4:

Input: nums = [10]
Output: "10"
 

Constraints:

1 <= nums.length <= 100
0 <= nums[i] <= 109

```python
class Solution:
    def largestNumber(self, nums: List[int]) -> str:
        """key point: compare str(a) +  str(b) with str(b) + str(a) to sort a, b"""
        """becareful with the case where nums consists of all 0's"""
        
        """Bubble Sort"""
        for i in range(len(nums)-1):
            for j in range(1, len(nums)):
                if str(nums[j-1]) + str(nums[j]) < str(nums[j]) + str(nums[j-1]):
                    nums[j-1], nums[j] = nums[j], nums[j-1]
        # res = []
        # for num in nums:
        #     res.append(str(num))
        # ans = ''.join(res)
        # return str(int(ans))
        return str(int("".join(map(str, nums)))) 
        """
        map(str, nums) converts nums to a list of string
        convert the string to int first to get rid of the leading 0's then convert back to string
        """
```

```python
class Solution:
    def largestNumber(self, nums: List[int]) -> str:
        """Insert Sort"""
        for i in range(1, len(nums)):
            tmp = nums[i]
            j = i - 1
            while j >= 0 and (str(tmp) + str(nums[j]) > str(nums[j]) + str(tmp)):
                nums[j+1] = nums[j]
                j -= 1
            nums[j+1] = tmp
        return str(int("".join(map(str, nums))))
```

```python
class Solution:
    def largestNumber(self, nums: List[int]) -> str:
        """Select Sort"""
        for i in range(len(nums)-1):
            max_loc = i
            for j in range(i+1, len(nums)):
                if str(nums[max_loc]) + str(nums[j]) < str(nums[j]) + str(nums[max_loc]):
                    max_loc = j
            if max_loc != i:
                nums[i], nums[max_loc] = nums[max_loc], nums[i]
        return str(int(''.join(map(str, nums))))
```

```python
class Solution:
    def largestNumber(self, nums: List[int]) -> str:
         """Merge Sort"""
        def merge(li, low, mid, high):
            i, j = low, mid + 1
            ltmp = []
            while i <= mid and j <= high:
                if str(li[i]) + str(li[j]) < str(li[j]) + str(li[i]):
                    ltmp.append(li[j])
                    j += 1
                else:
                    ltmp.append(li[i])
                    i += 1
            while i <= mid:
                ltmp.append(li[i])
                i += 1
            while j <= high:
                ltmp.append(li[j])
                j += 1
            li[low:high+1] = ltmp
        
        def merge_sort(li, low, high):
            if low < high:
                mid = (low + high)// 2
                merge_sort(li, low, mid)
                merge_sort(li, mid+1, high)
                merge(li, low, mid, high) 
        
        merge_sort(nums, 0, len(nums)-1)
        return str(int(''.join(map(str, nums))))
```

```python
class Solution:
    def largestNumber(self, nums: List[int]) -> str:
        """Quick Sort"""
        def partition(li, l, r):
            tmp = li[l]
            while l < r:
                # check from right
                while l < r and (str(tmp) + str(li[r]) >= str(li[r]) + str(tmp)):
                    r -= 1
                li[l], li[r] = li[r], li[l]
                # check from left
                while l < r and (str(tmp) + str(li[l]) <= str(li[l]) + str(tmp)):
                    l += 1
                li[l], li[r] = li[r], li[l]
            li[l] = tmp
            return l
        
        def quick_sort(li, l, r):
            if l < r:
                mid = partition(li, l, r)
                quick_sort(li, l, mid-1)
                quick_sort(li, mid+1, r)
                
        quick_sort(nums, 0, len(nums)-1)
        return str(int(''.join(map(str, nums))))
```
