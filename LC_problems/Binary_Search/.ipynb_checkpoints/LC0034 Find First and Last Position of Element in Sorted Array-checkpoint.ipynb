Given an array of integers nums sorted in ascending order, find the starting and ending position of a given target value.

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [5,7,7,8,8,10], target = 8
Output: [3,4]
Example 2:

Input: nums = [5,7,7,8,8,10], target = 6
Output: [-1,-1]
Example 3:

Input: nums = [], target = 0
Output: [-1,-1]
 

Constraints:

0 <= nums.length <= 105
-109 <= nums[i] <= 109
nums is a non-decreasing array.
-109 <= target <= 109
Accepted

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        """binary search"""
        # empty
        if not nums:
            return [-1, -1]
        # 1 num
        if len(nums) == 1:
            return [0, 0] if nums[0] == target else [-1, -1]
        left, right = 0, len(nums)-1
        while nums[left] < target or target < nums[right]:
            if nums[left] > target:
                return [-1, -1]
            if nums[right] < target:
                return [-1, -1]
            # nums[left] <= target <= [right]
            mid = left + (right - left)//2
            if nums[mid] < target:
                left = mid + 1
            elif nums[mid] > target:
                right = mid - 1
            # nums[mid] == target
            else:
                if nums[left] < target:
                    left += 1
                if nums[right] > target:
                    right -= 1
        return [left, right] if nums[left] == target else [-1, -1]
```

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        """binary search with recursion"""
        left, right = 0, len(nums)-1
        while left <= right:
            if nums[left] == nums[right] == target:
                return [left, right]
            mid = left + (right - left)//2
            if nums[mid] < target:
                left = mid + 1
            elif nums[mid] > target:
                right = mid - 1
            else:
                left = self.searchRange(nums[:mid+1], target)[0]
                right = self.searchRange(nums[mid+1:], target)[1]
                # if target not found in the left half, mid is the smallest index
                if left == -1:
                    return [mid, mid + 1 + right]
                # if target not found in the right hald, mid is the largest index
                if right == -1:
                    return [left, mid]
                return [left, mid + 1 + right]
        return [-1, -1]
```

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        """divide and conquer"""
        if not nums:
            return [-1, -1]
        def bs(l, r):
            if nums[l] == target == nums[r]:
                return [l, r]
            if nums[l] <= target <= nums[r]:
                m = l + (r-l)//2
                ll, rr = bs(l, m), bs(m+1, r)
                if -1 in ll:
                    return rr
                if -1 in rr:
                    return ll
                return [ll[0], rr[1]]
            return [-1, -1]
        return bs(0, len(nums)-1)
```

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        """2 binary searches"""
        # find the left end: the first index that makes nums[index] >= target
        def bsl():
            l, r = 0, len(nums)-1
            while l <= r:
                m = l + (r - l)//2
                if nums[m] < target: # index >= m doesn't satisy nums[index] >= target 
                    l = m + 1
                else:
                    r = m - 1
            return l
        # find the right end: the last index that makes nums[index] <= target
        def bsr():
            l, r = 0, len(nums)-1
            while l <= r:
                m = l + (r-l)//2
                if nums[m] > target: # index >= m doesn't satisfy nums[index] <= target
                    r = m - 1
                else:
                    l = m + 1
                # print('r', l, r)
            return r
        left, right = bsl(), bsr()
        # print(left, right)
        return [left, right] if left <= right else [-1, -1]
```

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]: 
        """2 binary seach: easier to understand"""
        def bsl():
            idx_l = -1
            l, r = 0, len(nums)-1
            while l <= r:
                m = l + (r - l)//2
                if nums[m] < target:
                    l = m + 1
                elif nums[m] > target:
                    r = m - 1
                else: # nums[m] == target
                    idx_l = m
                    r = m - 1 # search the left half
            return idx_l
        def bsr():
            idx_r = -1
            l, r = 0, len(nums)-1
            while l <= r:
                m = l + (r-l)//2
                if nums[m] < target:
                    l = m + 1
                elif nums[m] > target:
                    r = m - 1
                else: # nums[m] == target
                    idx_r = m
                    l = m + 1 # search the right half
            return idx_r
        return [bsl(), bsr()]
```
