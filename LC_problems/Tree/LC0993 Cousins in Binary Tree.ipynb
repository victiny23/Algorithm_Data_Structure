```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque
class Solution:
    def isCousins(self, root: TreeNode, x: int, y: int) -> bool:
        """level order to get the parent node value, depth for x and y, and store in a HashMap"""
        tree, queue = [], deque([(root, 0, None)]) # node, level, parent node value
        nodes = {}
        while queue:
            for i in range(len(queue)):
                node, depth, parent = queue.popleft()
                nodes[node.val] = [depth, parent]
                if node.left:
                    queue.append((node.left, depth + 1, node))
                if node.right:
                    queue.append((node.right, depth + 1, node))
        # check if x and y have the same level, and differetn aprents
        return nodes[x][0] == nodes[y][0] and nodes[x][1] != nodes[y][1]
```

```python
from collections import deque
class Solution:
    def isCousins(self, root: TreeNode, x: int, y: int) -> bool:
        """level order and group node values by level, 
        check if x and y in the same group, 
        early exit if x and y are brothers"""
        tree, queue = [], deque([root])
        while queue:
            level = []
            for i in range(len(queue)):
                node = queue.popleft()
                if node.left and node.right:
                    if (node.left.val == x and node.right.val == y) \
                    or (node.right.val == x and node.left.val == y):
                        return False
                level.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            tree.append(level)
            
        for depth in range(len(tree)):
            if x in tree[depth] and y in tree[depth]:
                return True
        return False
```

```python
from collections import deque
class Solution:
    def isCousins(self, root: TreeNode, x: int, y: int) -> bool:
        """recursive: depth-first search"""
        # search the nodes of value x or y, record the depth, parent
        def dfs(root, depth, parent, val):
            if not root:
                return
            # found the node of target value
            if root.val == val:
                return depth, parent
            # not found: search in left subtree and right subtree of the current node
            return dfs(root.left, depth+1, root, val) or dfs(root.right, depth+1, root, val)
        depth_x, parent_x = dfs(root, 0, None, x)
        depth_y, parent_y = dfs(root, 0, None, y)
        return depth_x == depth_y and parent_x != parent_y
```
