Given a binary tree, determine if it is height-balanced.

For this problem, a height-balanced binary tree is defined as:

a binary tree in which the left and right subtrees of every node differ in height by no more than 1.

 

Example 1:


Input: root = [3,9,20,null,null,15,7]
Output: true
Example 2:


Input: root = [1,2,2,3,3,null,null,4,4]
Output: false
Example 3:

Input: root = []
Output: true
 

Constraints:

The number of nodes in the tree is in the range [0, 5000].
-104 <= Node.val <= 104

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def depth(self, root):
        self.max_d = 0
        def dfs(root, depth): 
            if not root:
                self.max_d = 0 
                return
            if not root.left and not root.right:
                self.max_d = max(self.max_d, depth)
            if root.left:
                dfs(root.left, depth+1)
            if root.right:
                dfs(root.right, depth+1)
        dfs(root, 1)    
        return self.max_d

    def isBalanced(self, root: TreeNode) -> bool:
        """recusrion: check the difference of depth of subtrees"""
        if not root or (not root.left and not root.right):
            return True
        if not root.left:
            return not root.right.left and not root.right.right
        if not root.right:
            return not root.left.left and not root.left.right
        return self.isBalanced(root.left) and self.isBalanced(root.right) \
               and self.depth(root.left) - self.depth(root.right) in (-1, 0, 1)    
```

```python
class Solution:
    """recursion: simplified condition"""
    def getHeight(self, root):
        if not root:
            return 0
        return 1 + max(self.getHeight(root.left), self.getHeight(root.right))
    
    def isBalanced(self, root: TreeNode) -> bool:
        if not root:
            return True
        return self.isBalanced(root.left) and self.isBalanced(root.right) \
               and self.getHeight(root.left) - self.getHeight(root.right) in (-1, 0, 1)   
```

```python
class Solution:    
        """recursion: bottom-up comparion of depth"""
        def check(root):
            # check the max depth of the node
            if not root:
                return 0
            left = check(root.left)
            # return if left is not balanced before checking right saves time when the tree is skewed
            if left == -1:
                return -1
            right = check(root.right)
            if right == -1:
                return -1
            if abs(left - right) > 1:
                return -1
            return max(left, right) + 1
        
        return check(root) != -1
```
