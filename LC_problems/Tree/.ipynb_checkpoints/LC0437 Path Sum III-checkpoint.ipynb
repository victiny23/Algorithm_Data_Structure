Given the root of a binary tree and an integer targetSum, return the number of paths where the sum of the values along the path equals targetSum.

The path does not need to start or end at the root or a leaf, but it must go downwards (i.e., traveling only from parent nodes to child nodes).

 

Example 1:


Input: root = [10,5,-3,3,2,null,11,3,-2,null,1], targetSum = 8
Output: 3
Explanation: The paths that sum to 8 are shown.
Example 2:

Input: root = [5,4,8,11,null,13,4,7,2,null,null,5,1], targetSum = 22
Output: 3
 

Constraints:

The number of nodes in the tree is in the range [0, 1000].
-109 <= Node.val <= 109
-1000 <= targetSum <= 1000

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        """
        brutal: 
        1st layer: traversal through each node with any order
        2nd layer: check all the path sum (not necessarily ends on leaf)
        """
        self.cnt = 0
        # check all the path sum           
        def root_leaf_sum(root, targetSum):
            if not root:
                return 0
            cnt = 0
            stack = [(root, root.val)]
            while stack:
                node, curr_sum = stack.pop()
                if curr_sum == targetSum:
                    cnt += 1
                if node.left:
                    stack.append((node.left, curr_sum + node.left.val))
                if node.right:
                    stack.append((node.right, curr_sum + node.right.val))
            return cnt
        # traversal all the nodes   
        def dfs(root):
            if root:
                dfs(root.left)
                self.cnt += root_leaf_sum(root, targetSum)
                dfs(root.right)
        
        dfs(root)
        return self.cnt
```

```python
class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        """hash table: memorization of path sum"""
        self.cnt = 0
        # store the frequncy of sum to the current node from root
        cache = {0: 1}
        
        def dfs(root, prev_sum, target):
            if not root:
                return
            curr_sum = prev_sum + root.val
            if curr_sum - target in cache:
                self.cnt += cache[curr_sum - target] 
                # the number of old sum = the number of paths:
                # each node with an old sum equal to curr_sum - target 
                # provides a path from the child of that node to the current node
                # there could be more than 1 node with curr_sum equal to curr_sum - target
            cache[curr_sum] = cache.get(curr_sum, 0) + 1
            dfs(root.left, curr_sum, target)
            dfs(root.right, curr_sum, target)
            # moving from the parent node of the current node to a new branch
            # curr_sum from curr_node is no longer available
            cache[curr_sum] -= 1
            # no need to remove the key when value == 0, as 0 is added to the number of paths
            # if not cache[curr_sum]:
            #     cache.pop(curr_sum)
        
        dfs(root, 0, targetSum)
        return self.cnt
```
