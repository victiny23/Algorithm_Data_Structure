Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. (i.e., from left to right, then right to left for the next level and alternate between).

 

Example 1:


Input: root = [3,9,20,null,null,15,7]
Output: [[3],[20,9],[15,7]]
Example 2:

Input: root = [1]
Output: [[1]]
Example 3:

Input: root = []
Output: []
 

Constraints:

The number of nodes in the tree is in the range [0, 2000].
-100 <= Node.val <= 100

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque
class Solution:
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        """single ended queue + reverse"""
        if not root:
            return []
        tree, queue = [], deque([root])
        zigzag = 0
        # zigzag = -1
        while queue:
            level = []
            zigzag = 1 - zigzag
            # zigzag *= -1
            for i in range(len(queue)):
                node = queue.popleft()
                level.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            if zigzag:
                tree.append(level[::-1])
            else:
                tree.append(level)
            
            # tree.append(level[::zigzag])
            
            # tree.append(level if len(tree)%2 == 0 else level[::-1])
        return tree
```

```python
from collections import deque
class Solution:
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        """single ended queue + initializing array instead of reversing"""
        if not root:
            return []
        tree, queue = [], deque([root])
        even_level = False
        while queue:
            n = len(queue)
            level = [0] * n
            for i in range(n):
                node = queue.popleft()
                if even_level:
                    level[n-1-i] = node.val
                else:
                    level[i] = node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            even_level = not even_level
            tree.append(level)
        return tree
```

```python
from collections import deque
class Solution:
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        """double ended queue"""
        if not root:
            return []
        tree, queue = [], deque([root])
        even_level = False
        while queue:
            level = []
            if even_level:
                for i in range(len(queue)):
                    # pop node from the right of the queue
                    node = queue.pop()
                    level.append(node.val)
                    # appende nodes from the next level to the left of the queue
                    if node.right:
                        queue.appendleft(node.right)
                    if node.left:
                        queue.appendleft(node.left)
            else:
                for i in range(len(queue)):
                    # pop node from the left of the queue
                    node = queue.popleft()
                    level.append(node.val)
                    # appende nodes from the next level to the right of the queue
                    if node.left:
                        queue.append(node.left)
                    if node.right:
                        queue.append(node.right)
            even_level = not even_level
            tree.append(level)
        return tree
```

```python
from collections import deque
class Solution:
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        """deque for each level"""
        if not root:
            return []
        tree, queue = [], deque([root])
        even_level = False
        while queue:
            level = deque()
            for i in range(len(queue)):
                node = queue.popleft()
                if even_level:
                    level.appendleft(node.val)
                else:
                    level.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            even_level = not even_level
            tree.append(level)
        return tree
```
