You are given the root of a binary tree containing digits from 0 to 9 only.

Each root-to-leaf path in the tree represents a number.

For example, the root-to-leaf path 1 -> 2 -> 3 represents the number 123.
Return the total sum of all root-to-leaf numbers. Test cases are generated so that the answer will fit in a 32-bit integer.

A leaf node is a node with no children.

 

Example 1:


Input: root = [1,2,3]
Output: 25
Explanation:
The root-to-leaf path 1->2 represents the number 12.
The root-to-leaf path 1->3 represents the number 13.
Therefore, sum = 12 + 13 = 25.
Example 2:


Input: root = [4,9,0,5,1]
Output: 1026
Explanation:
The root-to-leaf path 4->9->5 represents the number 495.
The root-to-leaf path 4->9->1 represents the number 491.
The root-to-leaf path 4->0 represents the number 40.
Therefore, sum = 495 + 491 + 40 = 1026.
 

Constraints:

The number of nodes in the tree is in the range [1, 1000].
0 <= Node.val <= 9
The depth of the tree will not exceed 10.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque
class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        """recursion"""
        self.path_sum = 0
        def dfs(root):
            if not root:
                return 
            if not root.left and not root.right:
                self.path_sum += root.val
                return [1]     

            ldepths = dfs(root.left)
            if ldepths:
                for depth in ldepths:
                    self.path_sum += root.val*10**depth  

            rdepths = dfs(root.right)
            if rdepths:
                for depth in rdepths:
                    self.path_sum += root.val*10**depth

            if ldepths and rdepths:
                return [1 + d for d in ldepths + rdepths]
            if ldepths:
                return [1 + d for d in ldepths]
            if rdepths:
                return [1 + d for d in rdepths]
        dfs(root)
        return self.path_sum
```

```python
class Solution:
    def sumNumbers(self, root: TreeNode) -> int:    
        """iterative: dfs with stack"""
        path_sum = 0
        stack = [(root, root.val)]
        while stack:
            node, val =stack.pop()
            if not node.left and not node.right:
                path_sum += val
            if node.left:
                stack.append((node.left, val*10 + node.left.val))
            if node.right:
                stack.append((node.right, val*10 + node.right.val))
        return path_sum  
```

```python
class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        """iterative: bfs with deque"""
        path_sum = 0
        queue = deque([(root, root.val)])
        while queue:
            node, val = queue.popleft()
            if not node.left and not node.right:
                path_sum += val
            if node.left:
                queue.append((node.left, val*10 + node.left.val))
            if node.right:
                queue.append((node.right, val*10 + node.right.val))
        return path_sum
```
