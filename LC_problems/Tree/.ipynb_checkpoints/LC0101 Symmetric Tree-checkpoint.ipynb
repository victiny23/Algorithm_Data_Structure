Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

 

Example 1:


Input: root = [1,2,2,3,4,4,3]
Output: true
Example 2:


Input: root = [1,2,2,null,3,null,3]
Output: false
 

Constraints:

The number of nodes in the tree is in the range [1, 1000].
-100 <= Node.val <= 100

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque
class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        """recursive"""
        def isMirror(l, r):
            if not l and not r:
                return True
            if not l or not r:
                return False
            if l.val == r.val:
                return isMirror(l.left, r.right) and isMirror(l.right, r.left)
            return False
        
        if not root:
            return True
        return isMirror(root.left, root.right)
```

```python
from collections import deque
class Solution:
    def isSymmetric(self, root: TreeNode) -> bool: 
        """iterative: level order assign None"""
        """very slow with big extra sspace"""
        if not root:
            return True
        queue = deque([root])
        depth = 0
        while queue:
            """if all node in the queue is None"""
            if not any(queue):
                break
            n = len(queue)
            # assign a list of None for the level
            level = [None] * 2**depth
            depth += 1
            for i in range(n):
                node = queue.popleft()
                # node in current level
                if node:
                    # replace None with actual value
                    level[i] = node.val
                    # don't check if child node exists so place holder can be added
                    queue.append(node.left)
                    queue.append(node.right)
                # no node in current level, add place holder for the next level
                else:
                    queue.append(None)
                    queue.append(None)
            # check if the level is symmetric
            if level != level[::-1]:
                return False
        return True
        
        """improvement"""
        """iterative: level order by pair"""
        if not root:
            return True
        queue = deque([(root.left, root.right)]) # a deque[], each element is a pair ()
        while queue:
            l, r = queue.popleft()
            if not l and not r:
                continue
            if not l or not r:
                return False
            if l.val != r.val:
                return False
            queue.append((l.left, r.right))
            queue.append((l.right, r.left))
        return True
```
