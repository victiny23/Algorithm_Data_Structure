Given two integer arrays preorder and inorder where preorder is the preorder traversal of a binary tree and inorder is the inorder traversal of the same tree, construct and return the binary tree.

 

Example 1:


Input: preorder = [3,9,20,15,7], inorder = [9,3,15,20,7]
Output: [3,9,20,null,null,15,7]
Example 2:

Input: preorder = [-1], inorder = [-1]
Output: [-1]
 

Constraints:

1 <= preorder.length <= 3000
inorder.length == preorder.length
-3000 <= preorder[i], inorder[i] <= 3000
preorder and inorder consist of unique values.
Each value of inorder also appears in preorder.
preorder is guaranteed to be the preorder traversal of the tree.
inorder is guaranteed to be the inorder traversal of the tree.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        """recursive"""
        if not preorder:
            return None
        if len(inorder) == 1:
            return TreeNode(preorder[0])
        root = TreeNode(preorder[0])
        n = len(inorder)
        for i in range(n):
            if inorder[i] == preorder[0]:
                mid = i
                break
        root.left = self.buildTree(preorder[1:mid+1], inorder[:mid])
        root.right = self.buildTree(preorder[mid+1:], inorder[mid+1:])
        # return root
            
        """recursive: use index to locate the root"""
        if preorder:
            root = TreeNode(preorder[0])
            mid = inorder.index(preorder[0])
            root.left = self.buildTree(preorder[1:mid+1], inorder[:mid])
            root.right = self.buildTree(preorder[mid+1:], inorder[mid+1:])
            return root
        
        """recursive: use index to locate the root + update preorder list to save space"""
        if inorder: # this cannot be preorder
            mid = inorder.index(preorder.pop(0))
            root = TreeNode(inorder[mid])
            root.left = self.buildTree(preorder, inorder[:mid])
            root.right = self.buildTree(preorder, inorder[mid+1:])
            return root
        
        """recursive: no split, stop recursion at root.val"""
        def build(stop):
            if inorder and inorder[-1] != stop:
                root = TreeNode(preorder.pop())
                # left tree stops at the root.val
                root.left = build(root.val)
                # remove root.val from inorder
                inorder.pop()
                # right tree stops at the parent stopper: root.parent
                root.right = build(stop)
                return root
        # pop() is O(1) while pop(0) is O(n)
        preorder.reverse()
        inorder.reverse()
        return build(None)
```

```python
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        """Hash Table"""
        inorder.reverse()
        preorder.reverse()
        map_inorder = {}
        for i, val in enumerate(inorder):
            map_inorder[val] = i
        def recur(low, high):
            if low >= high:
                return
            node = TreeNode(preorder.pop())
            mid = map_inorder[node.val]
            node.left = recur(mid+1, high)
            node.right = recur(low, mid)
            return node
        return recur(0, len(inorder))
    
        """Hash Table: another edge case"""
        inorder.reverse()
        preorder.reverse()
        map_inorder = {}
        for i, val in enumerate(inorder):
            map_inorder[val] = i
        def recur(low, high):
            if low > high:
                return
            node = TreeNode(preorder.pop())
            mid = map_inorder[node.val]
            node.left = recur(mid+1, high)
            node.right = recur(low, mid-1)
            return node
        return recur(0, len(inorder)-1)
```
