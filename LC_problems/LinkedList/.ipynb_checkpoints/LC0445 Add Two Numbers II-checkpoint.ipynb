You are given two non-empty linked lists representing two non-negative integers. The most significant digit comes first and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

 

Example 1:


Input: l1 = [7,2,4,3], l2 = [5,6,4]
Output: [7,8,0,7]
Example 2:

Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [8,0,7]
Example 3:

Input: l1 = [0], l2 = [0]
Output: [0]
 

Constraints:

The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.
 

Follow up: Could you solve it without reversing the input lists?

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    """refer to LC0002 Add Two Numbers represented by reversed linked list"""
    def reverseList(self, head):
        if not head or not head.next:
            return head
        prev, curr = None, head
        while curr:
            foll = curr.next
            curr.next = prev
            prev = curr
            curr = foll
        return prev
            
    def reverse_addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)
        curr = dummy
        carry = 0
        while l1 or l2 or carry:
            if l1:
                carry += l1.val
                l1 = l1.next
            if l2:
                carry += l2.val
                l2 = l2.next
            curr.next = ListNode(carry%10)
            curr = curr.next
            carry //= 10
        return dummy.next
        
        
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        l1 = self.reverseList(l1)
        l2 = self.reverseList(l2)
        head  = self.reverse_addTwoNumbers(l1, l2)
        head = self.reverseList(head)
        return head
        
        x, y = 0, 0
        while l1:
            x = 10*x + l1.val
            l1 = l1.next
        while l2:
            y = 10*y + l2.val
            l2 = l2.next
        z = x + y
        dummy = ListNode(0)
        if not z:
            return dummy
        while z:
            v, z = z%10, z//10
            dummy.next, dummy.next.next = ListNode(v), dummy.next
        return dummy.next
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:  
        """use two stacks to store all the digits"""
        s1, s2 = [], []
        while l1:
            s1.append(l1.val)
            l1 = l1.next
        while l2:
            s2.append(l2.val)
            l2 = l2.next
            
        dummy = ListNode(0)
        carry = 0
        while s1 or s2 or carry:
            if s1:
                carry += s1.pop()
            if s2:
                carry += s2.pop()
            node = ListNode(carry%10)
            node.next = dummy.next
            dummy.next = node
            carry //= 10
        return dummy.next
```
