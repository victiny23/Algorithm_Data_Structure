Given the head of a linked list, return the list after sorting it in ascending order.

Follow up: Can you sort the linked list in O(n logn) time and O(1) memory (i.e. constant space)?

 

Example 1:


Input: head = [4,2,1,3]
Output: [1,2,3,4]
Example 2:


Input: head = [-1,5,3,4,0]
Output: [-1,0,3,4,5]
Example 3:

Input: head = []
Output: []
 

Constraints:

The number of nodes in the list is in the range [0, 5 * 104].
-105 <= Node.val <= 105

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:   
    """merge sort with O(nlogn)"""
    def sortList(self, head: ListNode) -> ListNode:
        if not head or not head.next:
            return head
        # find the mid point
        prev, slow, fast = None, head, head
        while fast and fast.next:
            prev, slow, fast = slow, slow.next, fast.next.next
        # break the linked list into two 
        prev.next = None
        # sort each part
        left = self.sortList(head)
        right = self.sortList(slow)
        # merge the two parts
        return self.merge(left, right)
    
    
    def merge(self, h1, h2):
        dummy = tail = ListNode(0)
        while h1 and h2:
            if h1.val < h2.val:
                tail.next, tail, h1 = h1, h1, h1.next
            else:
                tail.next, tail, h2 = h2, h2, h2.next
        tail.next = h1 or h2
        return dummy.next    
```

```python
class Solution:
    """quick sort with O(nlogn)"""
    """solved in ~60ms locally but TLE in submission for the same complicated testacse???"""
    def sortList(self, head: ListNode) -> ListNode:
        if not head or not head.next:
            return head
        left, mid, right = self.partition(head)
        left = self.sortList(left)
        right = self.sortList(right)
        q = mid
        while q.next:
            q = q.next
        q.next = right
        if not left: 
            return mid
        p = left
        while p.next:
            p = p.next
        p.next = mid
        return left
        
    # partition the linked list into two parts (<, >= mid value) + mid value   
    def partition(self, head):
        if not head or not head.next:
            return None, head, None
        right, p = head, head.next
        left = ListNode(0)
        left_tail = left
        right_tail = right
        while p:
            if p.val < right.val:
                right_tail.next = p.next
                left_tail.next = p
                left_tail = p
                p = right_tail.next
            else:
                right_tail, p = right_tail.next, p.next
        left_tail.next = None
        right_tail.next = None
        mid, right = right, right.next
        mid.next = None
        return left.next, mid, right
    
    # partition the linked list into three parts (<, =, > mid value)
    def partition(self, head):
        if not head or not head.next:
            return None, head, None
        left, mid, right = ListNode(0), ListNode(0), ListNode(0)
        l_tail, m_tail, r_tail = left, mid, right
        x = head.val
        while head:
            if head.val < x:
                l_tail.next = head
                l_tail = l_tail.next
            elif head.val > x:
                r_tail.next = head
                r_tail = r_tail.next
            else:
                m_tail.next = head
                m_tail = m_tail.next
            head = head.next
        l_tail.next, m_tail.next, r_tail.next = None, None, None
        return left.next, mid.next, right.next
```
