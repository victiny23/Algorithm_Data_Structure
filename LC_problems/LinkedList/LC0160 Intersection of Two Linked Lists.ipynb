Given the heads of two singly linked-lists headA and headB, return the node at which the two lists intersect. If the two linked lists have no intersection at all, return null.

It is guaranteed that there are no cycles anywhere in the entire linked structure.

Note that the linked lists must retain their original structure after the function returns.

 

Example 1:


Input: intersectVal = 8, listA = [4,1,8,4,5], listB = [5,6,1,8,4,5], skipA = 2, skipB = 3
Output: Intersected at '8'
Explanation: The intersected node's value is 8 (note that this must not be 0 if the two lists intersect).
From the head of A, it reads as [4,1,8,4,5]. From the head of B, it reads as [5,6,1,8,4,5]. There are 2 nodes before the intersected node in A; There are 3 nodes before the intersected node in B.
Example 2:


Input: intersectVal = 2, listA = [1,9,1,2,4], listB = [3,2,4], skipA = 3, skipB = 1
Output: Intersected at '2'
Explanation: The intersected node's value is 2 (note that this must not be 0 if the two lists intersect).
From the head of A, it reads as [1,9,1,2,4]. From the head of B, it reads as [3,2,4]. There are 3 nodes before the intersected node in A; There are 1 node before the intersected node in B.
Example 3:


Input: intersectVal = 0, listA = [2,6,4], listB = [1,5], skipA = 3, skipB = 2
Output: No intersection
Explanation: From the head of A, it reads as [2,6,4]. From the head of B, it reads as [1,5]. Since the two lists do not intersect, intersectVal must be 0, while skipA and skipB can be arbitrary values.
Explanation: The two lists do not intersect, so return null.
 

Constraints:

The number of nodes of listA is in the m.
The number of nodes of listB is in the n.
0 <= m, n <= 3 * 104
1 <= Node.val <= 105
0 <= skipA <= m
0 <= skipB <= n
intersectVal is 0 if listA and listB do not intersect.
intersectVal == listA[skipA + 1] == listB[skipB + 1] if listA and listB intersect.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:   
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        # add all the nodes of A to a Hash map
        # find if nodes of B exists in the Hash map
        a_nodes = {}
        while headA:
            a_nodes[headA] = 0
            headA = headA.next
        while headB:
            if headB in a_nodes:
                return headB
            else:
                headB = headB.next
        # no intersection
        return None
```

```python
class Solution: 
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:   
        # find length of both linked lists
        len1, len2 = 0, 0
        p1, p2 = headA, headB
        while p1:
            len1 += 1
            p1 = p1.next
        while p2:
            len2 += 1
            p2 = p2.next
        # reset pointer to the head
        p1, p2 = headA, headB
        # case 1
        if len1 > len2:
            len_dif = len1 - len2
            while len_dif > 0:
                p1 = p1.next
                len_dif -= 1
                # attention: it's wrong to use p1.data != p2.data
                # the values may be same while p1, and p2 pointing to different nodes
            while p1!= p2:
                p1 = p1.next
                p2 = p2.next
            return p1
        # case 2
        elif len1 < len2:
            len_dif = len2 - len1
            while len_dif > 0:
                p2 = p2.next
                len_dif -= 1
            while p1 != p2:
                p1 = p1.next
                p2 = p2.next
            return p1
        # case 3
        else:
            while p1 != p2:
                p1 = p1.next
                p2 = p2.next
            return p1
        # no intersection
        return None
```

```python
class Solution:
    def detectCycle(self, head):
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                while head != slow:
                    head = head.next
                    slow = slow.next
                return head
        return None
    
    
    
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        # connect the end of A to the head of A to create a cycle
        # from head of B detect the cycle
        # the intersection is at the entry of the cycle
        tailA = headA
        while tailA.next:
            tailA = tailA.next
        # create a cycle
        tailA.next = headA
        
        result = self.detectCycle(headB)
        # break the cycle
        tailA.next = None
        return result      
```
