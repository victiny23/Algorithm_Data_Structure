Given a linked list, reverse the nodes of a linked list k at a time and return its modified list.

k is a positive integer and is less than or equal to the length of the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should remain as it is.

You may not alter the values in the list's nodes, only nodes themselves may be changed.

 

Example 1:


Input: head = [1,2,3,4,5], k = 2
Output: [2,1,4,3,5]
Example 2:


Input: head = [1,2,3,4,5], k = 3
Output: [3,2,1,4,5]
Example 3:

Input: head = [1,2,3,4,5], k = 1
Output: [1,2,3,4,5]
Example 4:

Input: head = [1], k = 1
Output: [1]
 

Constraints:

The number of nodes in the list is in the range sz.
1 <= sz <= 5000
0 <= Node.val <= 1000
1 <= k <= sz

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    """recursive"""
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        curr = head
        g_len = k - 1
        # point to the last node of the group if node exists
        while curr and g_len > 0:
            curr = curr.next
            g_len -= 1
        # not enough nodes
        if g_len or not curr:
            return head
        # with enough nodes
        n_head = curr.next
        curr.next = None
        prev = head
        head = self.reverseList(head)
        prev.next = self.reverseKGroup(n_head, k)
        return head
  
    def reverseList(self, head):
        prev = None
        curr = head
        while curr:
            foll = curr.next
            curr.next = prev
            prev = curr
            curr = foll
        return prev
```

```python
class Solution: 
    """recursive"""
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        # point to the 1st node of the next group
        # 1
        curr = head
        cnt = 0
        while curr and cnt < k:
            curr = curr.next
            cnt += 1
        if cnt < k:
            return head
        
        # 2
        curr = head
        for _ in range(k):
            if not curr:
                return head
            curr = curr.next
            
        n_head, prev = self.reverse_n(head, k)
        head.next = self.reverseKGroup(n_head, k)
        return prev
    
    def reverse_n(self, head, n):
        prev, curr = None, head
        while n:
            foll = curr.next
            curr.next = prev
            prev = curr
            curr = foll
            n -= 1
        return (curr, prev)
        
```
