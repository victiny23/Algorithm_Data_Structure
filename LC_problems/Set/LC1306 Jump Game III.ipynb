Given an array of non-negative integers arr, you are initially positioned at start index of the array. When you are at index i, you can jump to i + arr[i] or i - arr[i], check if you can reach to any index with value 0.

Notice that you can not jump outside of the array at any time.

 

Example 1:

Input: arr = [4,2,3,0,3,1,2], start = 5
Output: true
Explanation: 
All possible ways to reach at index 3 with value 0 are: 
index 5 -> index 4 -> index 1 -> index 3 
index 5 -> index 6 -> index 4 -> index 1 -> index 3 
Example 2:

Input: arr = [4,2,3,0,3,1,2], start = 0
Output: true 
Explanation: 
One possible way to reach at index 3 with value 0 is: 
index 0 -> index 4 -> index 1 -> index 3
Example 3:

Input: arr = [3,0,2,1,2], start = 2
Output: false
Explanation: There is no way to reach at index 1 with value 0.
 

Constraints:

1 <= arr.length <= 5 * 104
0 <= arr[i] < arr.length
0 <= start < arr.length

```python
class Solution:
    def canReach(self, arr: List[int], start: int) -> bool:
         """DFS with Recursion: 469 ms + 76 MB"""
        seen  = [False] * len(arr)
        def dfs(nums, curr):
            if nums[curr] == 0:
                return True
            if curr - nums[curr] > -1 and not seen[curr - nums[curr]]:
                seen[curr - nums[curr]] = True
                left = dfs(nums, curr - nums[curr])
                if left:
                    return True
            if curr + nums[curr] < len(nums) and not seen[curr + nums[curr]]:
                seen[curr + nums[curr]] = True
                right = dfs(nums, curr + nums[curr])
                if right:
                    return True
            return False    
        return dfs(arr, start)
```

```python
class Solution:
    def canReach(self, arr: List[int], start: int) -> bool:
        """Recursion: 280 ms + 61 MB"""
        # check the bounds + check if the position has been visited
        if 0 <= start < len(arr) and arr[start] >= 0:
            arr[start] = -arr[start] # flip the checked number to mark it as visited
            return arr[start] == 0 or self.canReach(arr, start - arr[start]) \
                                   or self.canReach(arr, start + arr[start])
        return False
```

```python
class Solution:
    def canReach(self, arr: List[int], start: int) -> bool:
        """DFS with stack + set(): 310 ms + 21 MB"""
        stack, visited, n = [start], set(), len(arr)
        while stack:
            curr = stack.pop()
            if arr[curr] == 0:
                return True
            # mark curr as visited
            visited.add(curr)
            left = curr - arr[curr]
            if left >= 0 and left not in visited:
                stack.append(left)
            right = curr + arr[curr]
            if right < n and right not in visited:
                stack.append(right)
        return False
```

```python
class Solution:
    def canReach(self, arr: List[int], start: int) -> bool:
        """BFS with queue + set(): 336 ms + 21 MB"""
        from collections import deque
        q = deque([start])
        seen = set()
        n = len(arr)
        while q:
            curr = q.popleft()
            if arr[curr] == 0:
                return True
            seen.add(curr) # label curr as visited
            left = curr - arr[curr]
            if left >= 0 and left not in seen:
                q.append(left)
            right = curr + arr[curr]
            if right < n and right not in seen:
                q.append(right)
        return False
```
