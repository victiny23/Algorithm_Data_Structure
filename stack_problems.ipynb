```python
"""Balanced brackets"""
def isBalanced(s):
    dic = {')': '(', ']': '[', '}': "{"}
    stack = [s[0]]
    for i in range(1, len(s)):
        if s[i] not in dic.keys(): # left bracket
            stack.append(s[i])
        elif stack: # the stack is not empty
            if dic[s[i]] == stack[-1]: # right bracket matches the left bracket
                stack.pop()
            else: # not match
                stack.append(s[i])
        else: # the stack is empty, so the most left bracet is a right bracket -> not balanced
            return 'NO'  
    if stack: # extra bracket
        return 'NO'
    else:
        return 'YES'
```

```python
"""
Maximum element
append, pop, and find the maximum element in a stack
"""

n = int(input())
stack = []
mv = [0] 
for i in range(n):
    query = input().split(' ')
    cmd = int(query[0])
    if cmd == 1:
        val = int(query[1])
        stack.append(val)
        if i == 0:
            m = stack[-1]
            mv.append(m)
        else:
            if val >= m:
                m = val
                mv.append(m)
    elif cmd == 2:
        x = stack.pop()
        if x >= mv[-1]:
            mv.pop()
            m = mv[-1]
    else:
        print(m)

        
# a better solution -- only add new maximum to the stack
items = [0]
for i in range(int(input())):
    nums = list(map(int, input().split()))
    if nums[0] == 1:
        items.append(max(nums[1], items[-1]))
    elif nums[0] == 2:
        items.pop()
    else:
        print(items[-1])
```

```python
"""
Equal stacks

equalStacks has the following parameters:
int h1[n1]: the first array of heights
int h2[n2]: the second array of heights
int h3[n3]: the third array of heights

Returns
int: the height of the stacks when they are equalized
"""
def equalStacks(h1, h2, h3):
    h1 = h1[::-1]
    h2 = h2[::-1]
    h3 = h3[::-1]
    l1, l2, l3 = map(sum, (h1, h2, h3))
    
    while not l1 == l2 == l3:
        ml = min(l1, l2, l3)
        while l1 > ml:
            x1 = h1.pop()
            l1 -= x1
        while l2 > ml:
            x2 = h2.pop()
            l2 -= x2
        while l3 > ml:
            x3 = h3.pop()
            l3 -= x3
    return l1
```

```python
"""
Sum of two stacks

Alexa has two stacks of non-negative integers, a and b. Alexa challenges Nick to play the following game:

In each move, Nick can remove one integer from the top of either stack a or stack b.
Nick keeps a running sum of the integers he removes from the two stacks.
Nick is disqualified from the game if, at any point, his running sum becomes greater than some integer x given at the beginning of the game.
Nick's final score is the total number of integers he has removed from the two stacks.
Given a, b, and x, find the maximum possible score Nick can achieve 
(i.e., the maximum number of integers he can remove without being disqualified).
"""

"""
trap: a greedy algorithm of picking the smaller top element of the two stacks won't work

Consider two stacks s1 = [17,1,1,1,8] and s2 = [8,8,4,5,9] and max sum = 20. 
Using greedy approach you will select from s2 because 8 < 17. You will end up removing 3 ints from s2. 
Contrary to that if you pick from s1, you end up picking 4 ints
"""

def twoStacks(x, a, b):
    # reverse the list so the time complexity 
    # reduce from O(n) for a.pop(0) to O(1) for a.pop()
    a = a[::-1]
    b = b[::-1]
    
    total = 0
    # create a temp sum with element from only a
    # then compare the top elemenet of b and bottom element of a for optimization
    pa = []
    for i in range(len(a)):
        val1 = a.pop()
        if total + val1 > x:
            break
        total += val1
        pa.append(val1)
    
    max_cnt = len(pa)
    cur_cnt = max_cnt
    
    # we may check the top element of b adding to total, while possibly removing all the bottom element from a
    m = len(b)
    while len(b):
        # adding the top element of b to the temp sum
        # if exceed, remove the last added element from a and retry
        if total + b[-1] <= x:
            total += b.pop()
            cur_cnt += 1
            max_cnt = max(cur_cnt, max_cnt)
            continue
        # check on len(pa) must be here as the results may contain elements from only b
        # so we must compare total + b[-1] before break
        if not len(pa):
            break
        # removing the bottom element from a to update the total for possible addition with top element from b    
        val2 = pa.pop()
        total -= val2
        cur_cnt -= 1
        
    return max_cnt
```

```python
"""
Given a array of height of different blocks, each with width 1, find the rectangle with maximum area.
For example, the heights array [3, 2, 3]. A rectangle of height 2 and length 3 can be constructed within the boundaries.

Some examples:
h = [1, 3, 6, 8, 7, 2, 4, 1] yields 18 with h[2:5]
h = [8979, 4570, 6436, 5083, 7780, 3269, 5400, 7579, 2324, 2116] yields 26152 with h[0:8]

A one-direction extension strategy from left and right won't work because 
for some block with both sides higher than itself, only one side is considered and the other half is overlooked either from left or right
"""

def largestRectangle(h):
    ans = 0
    
    for i in range(len(h)):
        # find the central piece
        center = h[i]
        cnt = 1
        # extend to the left
        l = i - 1
        while l >= 0 and h[l] >= center:
            cnt += 1
            l -= 1
        # extend to the right
        r = i + 1
        while r < len(h) and h[r] >= center:
            cnt += 1
            r += 1
        # update result
        if cnt*center > ans:
            ans = cnt*center
    return ans
```
