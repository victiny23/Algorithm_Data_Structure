```python
"""Balanced brackets"""
def isBalanced(s):
    dic = {')': '(', ']': '[', '}': "{"}
    stack = [s[0]]
    for i in range(1, len(s)):
        if s[i] not in dic.keys(): # left bracket
            stack.append(s[i])
        elif stack: # the stack is not empty
            if dic[s[i]] == stack[-1]: # right bracket matches the left bracket
                stack.pop()
            else: # not match
                stack.append(s[i])
        else: # the stack is empty, so the most left bracet is a right bracket -> not balanced
            return 'NO'  
    if stack: # extra bracket
        return 'NO'
    else:
        return 'YES'
```

```python
"""
Maximum element
append, pop, and find the maximum element in a stack
"""

n = int(input())
stack = []
mv = [0] 
for i in range(n):
    query = input().split(' ')
    cmd = int(query[0])
    if cmd == 1:
        val = int(query[1])
        stack.append(val)
        if i == 0:
            m = stack[-1]
            mv.append(m)
        else:
            if val >= m:
                m = val
                mv.append(m)
    elif cmd == 2:
        x = stack.pop()
        if x >= mv[-1]:
            mv.pop()
            m = mv[-1]
    else:
        print(m)

        
# a better solution -- only add new maximum to the stack
items = [0]
for i in range(int(input())):
    nums = list(map(int, input().split()))
    if nums[0] == 1:
        items.append(max(nums[1], items[-1]))
    elif nums[0] == 2:
        items.pop()
    else:
        print(items[-1])
```

```python
"""
Equal stacks

equalStacks has the following parameters:
int h1[n1]: the first array of heights
int h2[n2]: the second array of heights
int h3[n3]: the third array of heights

Returns
int: the height of the stacks when they are equalized
"""
def equalStacks(h1, h2, h3):
    h1 = h1[::-1]
    h2 = h2[::-1]
    h3 = h3[::-1]
    l1, l2, l3 = map(sum, (h1, h2, h3))
    
    while not l1 == l2 == l3:
        ml = min(l1, l2, l3)
        while l1 > ml:
            x1 = h1.pop()
            l1 -= x1
        while l2 > ml:
            x2 = h2.pop()
            l2 -= x2
        while l3 > ml:
            x3 = h3.pop()
            l3 -= x3
    return l1
```
